```python

from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import bs4
import csv
from tqdm import tqdm_notebook as tqdm
import json
import pprint
```

```python
webpage = "https://www.tripadvisor.it/Hotels-g2440593-Province_of_Catania_Sicily-Hotels.html"
```

```python
response = requests.get(webpage)
```

```python
doc = bs4.BeautifulSoup(response.text)
```

```python
tipi = doc.find_all("span", class_="label" )


len(tipi)
```

```python

annunci = doc.find_all("div", class_="meta_listing" )


len(annunci)

print(annunci)
```

```python
annunci_list = []
for annuncio in annunci:
    cont_titolo=annuncio.find("div", class_="listing_title" )
    titolo = cont_titolo.find("a").text
    link = cont_titolo.find("a")["href"]
    review_count=annuncio.find("a", class_="review_count" ).text
    try: 
        tipologia_hotel=annuncio.find("span", class_="label" ).text
    except:
        pass       
    annunci_list.append({'titolo': titolo,'link':link,'tipologia_hotel':tipologia_hotel,'review_count':review_count})
print(annunci_list)
len(annunci_list)
```

```python
numero_annunci_pagina=int(len(annunci_list))
print(numero_annunci_pagina)
type (numero_annunci_pagina)
type(annunci_list)
```
